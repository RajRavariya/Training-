```python
def var_x(List):
    sum_sq=0
    for i in range(len(List)):
      sum_sq+=(List[i]-mean(List))**2
    var=sum_sq/(len(List)-1)
    sd=sqrt(var)
    return var,sd
    
```

```python
List=[1,2,3,4,5,6,7,8,9,12]
```

```python
var(List)*(10/9)
```

```python
var_x(List)
```

```python
def skewness(List):
    skewness=3*(mean(List)-median(List))/var_x(List)[1]
    return skewness
```

```python
skewness(List)
```
